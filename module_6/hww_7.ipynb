```python
class SingletonClass:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = super().__new__(cls) # створюється новий екземпляр класу, використовуючи поведінку методу __new__ за замовчуванням.
        return cls._instance  #super, a а не object  щоб гарантувати, що якщо клас є похідним від іншого класу, то для створення екземпляра буде використано відповідний базовий клас.


obj1 = SingletonClass()
obj2 = SingletonClass()
obj3 = SingletonClass()
obj4 = SingletonClass()

print(obj1 is obj3)  # Output: True
print(obj2 is obj4)  # Output: True
```

```python
import unittest


class Money:   #передаються аргументи evro та coins для ініціалізації атрибутів екземпляру self.amount та self.coins 
    def __init__(self, evro: int, coins: int):
        self.evro = evro
        self.coins = coins
                                #self посилається на екземпляр класу Money, на якому викликається метод.
    def __add__(self, other):   #other посилається на об'єкт, що додається до поточного екземпляру Money.
        if isinstance(other, Money):    # перевіряє, чи other є екземпляром класу Money - int
            total_evro = self.evro + other.evro
            total_coins = self.coins + other.coins
            if total_coins >= 100:  #  
                total_evro += 1
                total_coins -= 100
            return Money(total_evro, total_coins)
        else:
            raise TypeError("Unsupported operand type for +")
            


    def __sub__(self, other):
        if isinstance(other, Money):
            if self.evro > other.evro or (self.evro == other.evro and self.coins >= other.coins):
                total_evro = self.evro - other.evro
                total_coins = self.coins - other.coins
                if total_coins < 0:
                    total_evro -= 1
                    total_coins += 100
                return Money(total_evro, total_coins)
            else:  #Якщо віднімання не є дійсним (тобто self є меншим за other)
                raise ValueError("Cannot subtract a larger evro")
        else:
            raise TypeError("Unsupported operand type for -")

    def __eq__(self, other):
        if isinstance(other, Money):
            return self.evro == other.evro and self.coins == other.coins
        return False

    def __ne__(self, other):
        return not self.__eq__(other)

    def __repr__(self):  #  повертає рядок (16, 10), який можна використати для відтворення об'єкта
        return f"Money({self.evro}, {self.coins})"



class MoneyTests(unittest.TestCase):
    def test_addition(self):
        money1 = Money(10, 50)
        money2 = Money(5, 60)
        expected_result = Money(16, 10) 
        #оператор ствердження, який перевіряє,чи операція додавання(money1 + money2) дає очікуваний результат(expected_result
        self.assertEqual(money1 + money2, expected_result)

    def test_subtraction(self):
        money1 = Money(10, 50)
        money2 = Money(5, 60)
        expected_result = Money(4, 90)
        self.assertEqual(money1 - money2, expected_result)

    def test_equality(self):
        money1 = Money(10, 50)
        money2 = Money(10, 50)
        self.assertEqual(money1, money2)

    def test_inequality(self):
        money1 = Money(10, 50)
        money2 = Money(5, 60)
        self.assertNotEqual(money1, money2)

# if __name__ == "__main__":
#         unittest.main()

#loadTestsFromTestCase метод з класу TestLoader, який автоматично знаходить і завантажує всі тестові методи
suite = unittest.TestLoader().loadTestsFromTestCase(MoneyTests)

# створення екземпляра тестового бігуна, який відповідає за виконання тестів і відображення результатів
runner = unittest.TextTestRunner()
runner.run(suite)  #використовується для виконання набору(suite) тестів за допомогою тестового бігуна
```

```python
# використовується для створення унікального ідентифікатора для кожного банківського рахунку. Цей унікальний 
# ідентифікатор присвоюється атрибуту id екземпляру BankAccount
import uuid 
import logging


class ContactInfo: # init конструктором, який виконується, коли створюється новий екземпляр клас
    def __init__(self, name, address, email, phone):
        self.name = name
        self.address = address
        self.email = email
        self.phone = phone

class Money:
    def __init__(self, evro, coins):
        self.evro = evro
        self.coins = coins

#  клас BankAccount, ви можете створювати екземпляри цього класу, що представляють окремі банківські рахунки. Кожен 
#рахунок має унікальний ідентифікатор, контактну інформацію, баланс і логгер для реєстрації дій, пов'язаних з рахунком
class BankAccount:
    def __init__(self, contact_info, initial_balance):
        self.id = uuid.uuid4().hex  # генерує унікальний ідентифікатор банківського рахунку за допомогою модуля uuid. 
        self.contact_info = contact_info  #Функція uuid4() генерує випадковий UUID, а .hex перетворює його в шістнадцяткове представлення рядка.
        self.balance = Money(initial_balance.evro, initial_balance.coins) #створює об'єкт Money, що представляє початковий баланс рахунку initial_balance
        self.logger = logging.getLogger('BankAccount') #дозволяє записувати повідомлення, пов'язані з рахунком
        self.logger.setLevel(logging.INFO) #до журналу будуть записуватися лише повідомлення з рівнем INFO і вище
        self.logger.addHandler(logging.StreamHandler()) #додає обробник потоку до журналу. StreamHandler відповідає за виведення повідомлень журналу на консоль.

    

    def __str__(self):  #відповідає за створення рядкового представлення екземпляру BankAccount
        return f"BankAccount({self.id}): {self.contact_info.name}, Address: {self.contact_info.address}, Email: {self.contact_info.email}, Phone: {self.contact_info.phone}, Balance: {self.balance.evro}.{self.balance.coins}"

    
    def __repr__(self):
        return f"BankAccount(id='{self.id}', contact_info={repr(self.contact_info)}, balance={repr(self.balance)})"

    def replenish(self, money):
        self.balance.evro += money.evro
        self.balance.coins += money.coins #використовується для запису інформаційного повідомлення про операцію поповнення
        if self.balance.coins >= 100:
            self.balance.evro += 1
            self.balance.coins -= 100
        self.logger.info(f"Replenish {money.evro}.{money.coins} to BankAccount {self.id}")
        
    def withdraw(self, money):
        if self.balance.evro > money.evro or (self.balance.evro == money.evro and self.balance.coins >= money.coins):
            self.balance.evro -= money.evro
            self.balance.coins -= money.coins
            if self.balance.coins < 0:
                self.balance.evro -= 1
                self.balance.coins += 100
            self.logger.info(f"Withdrew {money.evro}.{money.coins} from BankAccount {self.id}")
        else:
            self.logger.error(f"Withdrawal failed: Insufficient balance in BankAccount {self.id}")

if __name__ == "__main__":

    contact_info = ContactInfo("Will Smit", "Tarasa Shevchenka 43", "w.Smit@example.com", "0675439978")
    bank_account = BankAccount(contact_info, Money(100, 32))
    bank_account.replenish(Money(500099000, 25))
    bank_account.withdraw(Money(39324799, 75))
    print(bank_account)
    print(bank_account.__repr__)
```

```python
import sqlite3
 
class define_table:    #інкапсулює функціональність ORM
    def __init__(self, db_name):   #db_name представляє собою ім'я або шлях до файлу бази даних SQLite
        self.conn = sqlite3.connect(db_name)  #встановлює з'єднання з базою даних SQLite, створюється об'єкт з'єднання db_name
        self.cursor = self.conn.cursor()  #Об'єкт курсору дозволяє виконувати SQL-оператори та отримувати результати запитіd

    def close(self):
        self.conn.close()
        
    def create_table(self, table_name, columns): #
        columns_str = ', '.join(columns) #об'єднує назви стовпців і типи даних зі списку стовпців у рядок, розділений комами
        create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({columns_str})"#формування SQL-запит для створення таблицi
        self.cursor.execute(create_table_query) #виконує SQL-запит(створення). дійсний об'єкт курсору для виконання операторів SQL
        self.conn.commit()

    def find_by_key(self, table_name, key): #self (представляє екземпляр класу SimpleORM)
        find_query = f"SELECT * FROM {table_name} WHERE id = ?" #конструює SQL-запит на вибірку всіх стовпців (*) з вказаного table_name і застосовує умову за допомогою речення WHERE для фільтрації рядків, де стовпець id дорівнює параметризованому значенню (?
        self.cursor.execute(find_query, (key,)) #Значення параметра key надається у вигляді одноелементного кортежу (key,) методу execute для заміни параметризованого значення (?) у запитi find_query
        result = self.cursor.fetchone()  #повертає перший рядок з результату запиту
        return result  #Якщо не знайдено жодного відповідного рядка, результатом буде None

    def take_all(self, table_name): #отримання всіх рядків з певної таблиці в базі даних SQLite
        take_all_query = f"SELECT * FROM {table_name}" #вибірку всіх стовпців (*) з вказаного імені таблиці
        self.cursor.execute(take_all_query) #надсилає запит до рушія бази даних для виконання
        result = self.cursor.fetchall() #отримує всі рядки, повернуті запитом, у вигляді списку кортежі
        return result

    def insert(self, table_name, data): #data (словник, що представляє назви стовпців і відповідні їм значення для нового рядка)
        columns = ', '.join(data.keys()) #об'єднує ключі словника даних(представляють назви стовпців)у рядок, розділений комами
        placeholders = ', '.join(['?' for _ in data]) #створює рядок заповнювачів (?)id, для кожного стовпця у словнику даниx
        values = tuple(data.values()) #отримує значення зі словника даних і перетворює їх у кортеж
        
        #SQL-запит для вставки нового рядка в таблицю (заповнювачі)
        insert_query = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
        self.cursor.execute(insert_query, values) #замінює пропуски (?) у запиті відповідними значеннями з кортежу values
        self.conn.commit()  #гарантує, що вставлений рядок буде постійно зберігатися в базі даних

    def update(self, table_name, key, data): #cтворює рядок присвоєнь стовпців у форматі column = ? 
        set_values = ', '.join([f"{column} = ?" for column in data.keys()]) #для кожного стовпця у словнику даних
        values = tuple(data.values()) #отримує нові значення зі словника даних і перетворює їх у кортеж.
        
        #f-рядок поєднує ім'я_таблиці, set_values та умову для визначення рядка
        update_query = f"UPDATE {table_name} SET {set_values} WHERE id = ?"
        self.cursor.execute(update_query, values + (key,)) #execute замінює заповнювачі (?) у запиті відповідними значеннями з кортежу values. key додається, щоб вказати рядок, який потрібно оновити на основі стовпця id.
        self.conn.commit()

    def delete(self, table_name, key):
        delete_query = f"DELETE FROM {table_name} WHERE id = ?" #SQL-запит на видалення рядка з вказаного імені_таблиц
        self.cursor.execute(delete_query, (key,)) #одноелементного кортежу (key,)
        self.conn.commit()

db = define_table("hw74.db")

db.create_table("users", ["id INTEGER PRIMARY KEY", "name TEXT", "age INTEGER"])

db.insert("users", {"name": "kate", "age": 35})

db.update("users", 1, {"age": 46})

result = db.find_by_key("users", 1)
print(result)  

all_data = db.take_all("users")
print(all_data)  

# db.delete("users", 1)
db.close()
```
